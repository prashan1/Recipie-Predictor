```python
#sns.lmplot('Flour' , 'Sugar', data = recipies , hue = 'Type', palette = 'Set1', fit_reg=False, scatter_kws={"s": 70})
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from  sklearn import metrics
import seaborn as sns
```

```python
# Fetching dataset Online
recipie = pd.read_csv('https://raw.githubusercontent.com/adashofdata/muffin-cupcake/master/recipes_muffins_cupcakes.csv')
recipie
```

```python
X = recipie.loc[:, ['Flour' , 'Sugar'] ]
y = recipie['Type']
```

```python
sns.lmplot('Flour' ,'Sugar' ,data = recipie, fit_reg = False , palette = 'Set1', hue = 'Type',scatter_kws ={'s': 70} )
```

```python
from sklearn import svm , metrics
from sklearn.model_selection import train_test_split
clf = svm.SVC(kernel = 'linear' )
X_train , X_test , y_train ,y_test = train_test_split(X , y )
clf.fit(X, y )

# Get the separating hyperplane
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(30, 60)
yy = a * xx - (clf.intercept_[0]) / w[1]

# Plot the parallels to the separating hyperplane that pass through the support vectors
b = clf.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = clf.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])
sns.lmplot('Flour', 'Sugar', data=recipie, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black');
```

```python

# Look at the margins and support vectors
sns.lmplot('Flour', 'Sugar', data=recipie, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
```

```python
sns.lmplot('Flour', 'Sugar', data=recipie, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
prict = clf.predict([[45,15]])
plt.plot([32],[15] , 'ko')
plt.show()
```
